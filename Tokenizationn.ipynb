# Tokenization Example

```python
corpus=''' Hello welcome ,to yash shahi's NLP Walkthrough.
please look around and find out !  to become an expert in NLP'''
```

```python
print(corpus)
```

```python
#setting up to use tokenize for the first time 
import nltk
nltk.download("punkt")
```

```python
print(nltk.data.path)
```

```python
nltk.download('punkt_tab')
```

```python
## tokenization 
#sentence ---> paragraphs
from nltk.tokenize import sent_tokenize
```

```python
sent_tokenize(corpus)
```

```python
documents=sent_tokenize(corpus)
```

```python
type(documents)
```

```python
for sentence in documents:
    print(sentence)
```

```python
#tokenization
#Paragraph -->words
#sentence -->words

from nltk.tokenize import word_tokenize
```

```python
word_tokenize(corpus)
#, . are treated as seprate word
```

```python
#how to do tokenizaiton on sentence
for sentence in documents:
    print(word_tokenize(sentence))
```

```python
from nltk.tokenize import wordpunct_tokenize
```

```python
wordpunct_tokenize(corpus)

#shahi's --> shahi,',s the aporstrphe got splited
```

```python
from nltk.tokenize import TreebankWordDetokenizer
```

```python
tokenizer=TreebankWordDetokenizer()
#initializing this
```

```python
tokenizer.tokenize(corpus)
#full stop wont be treated as a seprate word but with respect to last word full stop will be seprate
```
