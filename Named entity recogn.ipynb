### **Named Entity Recognition**

```python
sentence="The Eiffel Tower was built from 1887 to 1889 by Gustave Eiffel, whose company specialized in building metal frameworks and structures."
```

```python
import nltk
words=nltk.word_tokenize(sentence)
```

```python
tag_elements=nltk.pos_tag(words)
```

```python
nltk.download('maxent_ne_chunker_tab')
```

```python
nltk.download('words')
```

```python
nltk.ne_chunk(tag_elements).draw()
```

Named Entity Recognition (NER) is important because it **automatically identifies proper nouns** like people, organizations, and locations in text. Using NLTK, it helps **structure unstructured text**, making it easier to analyze and extract meaningful information. NER is **crucial for tasks like information retrieval, question answering, and text summarization**.
