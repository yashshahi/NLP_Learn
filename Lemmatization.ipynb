# **Wordnet Lemmatizer**

lemmatization technique is like stemming .The output we will get after lemmatization is called 'lemma', which is a root word rather than a root stem, the output of stemming. after lemmatization,we will be getting a valid word that means the same thing.

NLTK provides WordNetLemmatizer class which is a thin wrapper around the wordnet corpus .this class use morphy() function  to the WordNet CorpusReader class to find a lemma 

```python
import nltk
```

```python
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer=WordNetLemmatizer()
```

```python
'''
POS- Noun-n 
verb-v
adjective-a
adverb-r

by default postag is "n"
'''
lemmatizer.lemmatize("going",pos="n")
```

```python
lemmatizer.lemmatize("going",pos="v")
```

```python
words=['eating',"eat", "eaten","writing","write","programing","programs","history","finally ","finalize"]
```

```python
for word in words:
    print(word+"---->"+lemmatizer.lemmatize(word,pos='v'))
```
